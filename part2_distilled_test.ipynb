```python
convert = []
mileage_type = []

for i in df1.index:
    if df1.loc[i, 'mileageType'] == 'miles':
        mileage_type.append('kilometres')        
        kilometers = int(df1.loc[i, 'mileage'] * 1.609344)
        convert.append(kilometers)
    else:
        mileage_type.append('kilometres')
        convert.append(df1.loc[i, 'mileage'])
        
```

```python
df1['kilometers'] = convert
df1['type'] = mileage_type
```
